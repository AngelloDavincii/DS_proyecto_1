```python
import pandas as pd 
import glob 
import numpy as np 
import matplotlib.pyplot as plt 
import numpy as np
```

<h2><center>Departamento de Guatemala</center></h2>

```python
path_prefix = "data_xlsx/GUATEMALA/"
datos_en_bruto = glob.glob(path_prefix+"*.xlsx")
paths = []
datas = {}
for i in datos_en_bruto:
    a= i.split("\\")[-1]
    paths.append(a)
    
data_frames_separados_guatemala = []
for j in paths:
    df = pd.read_excel(j)
    data_frames_separados_guatemala.append(df)

    
guatemala_departamento = pd.concat(data_frames_separados_guatemala)
datas["Guatemala"] = guatemala_departamento
```

```python
from pandas_profiling import ProfileReport
```

```python
prof = ProfileReport(guatemala_departamento)
prof
```

<h2><center>Todos los departamentos</center></h2>

```python
path_prefix = "data_xlsx/"
datos_en_bruto_globales = glob.glob("data_xlsx/*.xlsx")
paths_globales = []
for i in datos_en_bruto_globales:
    a= i.split("\\")[-1]
    paths_globales.append(a)

departamentos = []
for j in paths_globales :
    df = pd.read_excel(j)
    departamentos.append(df)
    datas[j.replace(path_prefix,"").replace(".xlsx", "")] = df
    
departamentos.append(guatemala_departamento)
datas
```

```python
dataframe_pais = pd.concat(departamentos)
dataframe_pais
```
